```python
#functie voor isbn 10 en 13 
def is_isbn(isbn):
    isbn = isbn.replace('-', '').replace(' ', '')  # Verwijder streepjes en spaties
    if len(isbn) == 10 and isbn[:-1].isdigit() and (isbn[-1].isdigit() or isbn[-1].upper() == 'X'):
        total = sum((10 - i) * (int(x) if x != 'X' else 10) for i, x in enumerate(isbn))
        return total % 11 == 0
    elif len(isbn) == 13 and isbn.isdigit() and (isbn.startswith('978') or isbn.startswith('979')):
        total = sum((int(x) * 3 if i % 2 else int(x)) for i, x in enumerate(isbn[:-1]))
        checksum = 10 - (total % 10)
        return str(checksum % 10) == isbn[-1]
    else:
        return False
```

```python
# toepassen op kolom 
Ratings['is_isbn'] = Ratings['Id'].astype(str).apply(is_isbn)

# hoeveel rijen zijn ISBN?
print( Ratings['is_isbn'].sum())
```

```python

invalid_count = 0
print("ongeldige ISBN nummers:")
for index, row in Ratings.iterrows():
    if not is_isbn(row['Id']):
        print(row['Id'])
        invalid_count += 1
        if invalid_count >= 20:  # max aantal anders print die alles 
            break
```

```python
#alleen rijen behouden met ISBN
rating_isbn = Ratings[Ratings['Id'].astype(str).apply(is_isbn)]
```

```python
# naam veranderen naar ISBN
rating_isbn = rating_isbn.rename(columns={'Id': 'ISBN'})

rating_isbn.head()
```

```python
rating_isbn = rating_isbn.dropna(subset=['Title', 'profileName', 'User_id', 'Title', 'review/summary', 'review/text'])


print(rating_isbn.isna().sum())
```

```python
# Verwijder onnodige kolommen
rating_isbn = rating_isbn.drop(columns=['User_id','Price', 'is_isbn'])

rating_isbn.head(15)
```

```python
rating_subset_1k = rating_isbn.sample(n=1000, random_state=42)
rating_subset_1k.info()
```

```python

rating_subset_1k.reset_index(drop=True, inplace=True)

rating_subset_1k
```

```python

def score_to_text(score):
    if score < 1.5:
        return "Slecht"
    elif score < 2.5:
        return "Matig"
    elif score < 3.5:
        return "Gemiddeld"
    elif score < 4.5:
        return "Goed"
    else:
        return "Uitstekend"


rating_subset_1k['text_rating'] = rating_subset_1k['review/score'].apply(score_to_text)


rating_subset_1k.head()
```
